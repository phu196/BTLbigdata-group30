```python
# Cell 1
import pandas as pd
import numpy as np
from faker import Faker
import random, uuid, json
from datetime import datetime, timedelta
from tqdm import tqdm

fake = Faker("vi_VN")
Faker_seed = 42
random.seed(Faker_seed)
np.random.seed(Faker_seed)
```

```python
# Cell 2 - cấu hình quy mô
NUM_STUDENTS = 3000      # số sinh viên (tăng lên nếu muốn)
NUM_TEACHERS = 20        # số giảng viên
NUM_COURSES = 40         # số môn học
SEMESTERS = ["2023-1","2023-2","2024-1","2024-2"]  # các học kỳ có dữ liệu lịch sử
MAX_CLASSES_PER_SEM = 20 # lớp học phần tồn tại mỗi học kỳ
```

```python
# Cell 3 - students
students = []
for i in range(NUM_STUDENTS):
    sid = f"S{20220000 + i}"
    dob = fake.date_of_birth(minimum_age=21, maximum_age=22)
    students.append({
        "student_id": sid,
        "full_name": fake.name(),
        "dob": dob.isoformat(),
        "faculty": random.choice(["IT1","IT2","IT-E10","IT-E7","IT-E6"]),
        "email": f"{sid.lower()}@hust.edu.vn",
        "password": fake.password(length=10)  # giả lập (không dùng thật)
    })

df_students = pd.DataFrame(students)
df_students.head()
```

```python
# Cell 4 - teachers
teachers = []
for i in range(NUM_TEACHERS):
    tid = f"T{1000 + i}"
    teachers.append({
        "teacher_id": tid,
        "full_name": fake.name(),
        "specialty": random.choice(["Data Science","DB","AI","Networks","Software Engineering","Mathematics"]),
        "email": f"{tid.lower()}@university.edu.vn",
        "password": fake.password(10)
    })

df_teachers = pd.DataFrame(teachers)
df_teachers.head()
```

```python
# Cell 5 - courses
subjects_base = [
    "Programming 1","Data Structures","Databases","Algorithms","Operating Systems",
    "Computer Networks","Machine Learning","Big Data","Statistics","Software Engineering",
    "Calculus","Linear Algebra","Digital Logic","Probability","Web Development"
]
# mở rộng số môn nếu cần
courses = []
for i in range(NUM_COURSES):
    cid = f"C{200 + i}"
    name = random.choice(subjects_base) + (f" ({i//len(subjects_base)+1})" if NUM_COURSES>len(subjects_base) else "")
    courses.append({
        "course_id": cid,
        "course_name": name,
        "credits": random.choice([2,3,4]),
        "teacher_id": random.choice(df_teachers["teacher_id"].tolist())
    })

df_courses = pd.DataFrame(courses)
df_courses.head()
```

```python
# Cell 6 - class sections
class_sections = []
class_id_counter = 1
for sem in SEMESTERS:
    for _ in range(MAX_CLASSES_PER_SEM):
        course = random.choice(df_courses.to_dict("records"))
        class_id = f"CL{class_id_counter:04d}"
        class_id_counter += 1
        # random số sinh viên đăng ký mỗi lớp
        capacity = random.randint(10, 80)
        class_sections.append({
            "class_id": class_id,
            "course_id": course["course_id"],
            "semester": sem,
            "year": sem.split("-")[0],
            "teacher_id": course["teacher_id"],
            "capacity": capacity
        })

df_classes = pd.DataFrame(class_sections)
df_classes.head()
```

```python
# Cell 7 - enrollments (một sinh viên có thể đăng ký nhiều môn)
enrollments = []
student_ids = df_students["student_id"].tolist()
class_ids = df_classes["class_id"].tolist()

# đảm bảo phân bổ hợp lý: mỗi student 4-7 môn mỗi học kỳ (khi có)
for sem in SEMESTERS:
    classes_in_sem = df_classes[df_classes["semester"]==sem]["class_id"].tolist()
    for sid in student_ids:
        n = random.randint(3,8)  # số môn đăng ký
        chosen = random.sample(classes_in_sem, min(n, len(classes_in_sem)))
        for cl in chosen:
            enrollments.append({
                "student_id": sid,
                "class_id": cl,
                "semester": sem,
                "enroll_date": fake.date_between(start_date="-365d", end_date="today").isoformat()
            })

df_enroll = pd.DataFrame(enrollments).drop_duplicates(subset=["student_id","class_id"])
df_enroll.head()
```

```python
# Cell 8 - sessions and attendance per class per semester
sessions = []   # từng buổi học với session_id
attendance = [] # attendance entries

def mk_dates_for_semester(semester_label, start_date=None, weeks=12):
    # trả về danh sách ngày học (tuần x 2 buổi) giả lập
    if start_date is None:
        # đơn giản: chọn ngày bắt đầu theo semester
        year = int(semester_label.split("-")[0])
        start_date = datetime(year, 2, 1) if "1" in semester_label else datetime(year, 9, 1)
    ds = []
    for w in range(weeks):
        # mỗi tuần 2 buổi: thứ 2 và thứ 4 giả sử
        d1 = start_date + timedelta(days=w*7)
        d2 = d1 + timedelta(days=2)
        ds.append(d1.date().isoformat())
        ds.append(d2.date().isoformat())
    return ds

session_id_counter = 1
for idx, row in df_classes.iterrows():
    sem = row["semester"]
    dates = mk_dates_for_semester(sem, weeks=10)
    for d in dates:
        sid = f"SES{session_id_counter:06d}"
        session_id_counter += 1
        sessions.append({
            "session_id": sid,
            "class_id": row["class_id"],
            "date": d
        })
        # Attendance: cho những student đăng ký lớp đó
        enrolled = df_enroll[df_enroll["class_id"]==row["class_id"]]["student_id"].tolist()
        for stu in enrolled:
            # xác suất vắng có phép / vắng không phép
            r = random.random()
            if r < 0.85:
                status = "present"
            elif r < 0.92:
                status = "excused"
            else:
                status = "absent"
            attendance.append({
                "session_id": sid,
                "student_id": stu,
                "status": status
            })

df_sessions = pd.DataFrame(sessions)
df_attendance = pd.DataFrame(attendance)
df_sessions.head(), df_attendance.head()
```

```python
# Cell 9 - grades generation with weights
grades = []
# weights có thể khác nhau theo lớp
for cl in df_classes["class_id"].tolist():
    enrolled_students = df_enroll[df_enroll["class_id"]==cl]["student_id"].tolist()
    weight_mid = random.choice([0.3, 0.4])
    weight_final = 1 - weight_mid
    for sid in enrolled_students:
        mid = round(max(0, min(10, np.random.normal(loc=6.5, scale=1.8))),2)
        final = round(max(0, min(10, np.random.normal(loc=6.3, scale=2.0))),2)
        total = round(mid * weight_mid + final * weight_final, 2)
        grades.append({
            "student_id": sid,
            "class_id": cl,
            "semester": df_classes[df_classes["class_id"]==cl]["semester"].values[0],
            "midterm_score": mid,
            "final_score": final,
            "weight_mid": weight_mid,
            "weight_final": weight_final,
            "total_score": total,
            "passed": mid >= 3.0 and final >= 3.0 and total >= 4.0
        })

df_grades = pd.DataFrame(grades)
df_grades.head()
```

```python
# Cell 10 - generate streaming-like activity logs (tạo file lớn hoặc stream)
activities_list = []
event_id = 1
activity_types = ["login","view_video","submit_assignment","submit_quiz","forum_post","download_material"]
# thời gian giả sử trong 30 ngày gần đây
now = datetime.now()
start_ts = now - timedelta(days=30)

# Tạo N events (tăng N để mô phỏng volume lớn)
N_EVENTS = NUM_STUDENTS * 50  # mỗi sv ~50 event trong 30 ngày

student_ids = df_students["student_id"].tolist()

for i in tqdm(range(N_EVENTS)):
    stu = random.choice(student_ids)
    cl = random.choice(df_enroll[df_enroll["student_id"]==stu]["class_id"].tolist())
    ts = start_ts + timedelta(seconds=random.randint(0, 30*24*3600))
    act = random.choice(activity_types)
    duration = random.randint(10, 3600) if act in ["view_video","submit_assignment"] else None
    score = round(random.uniform(0,10),2) if act in ["submit_quiz","submit_assignment"] and random.random()>0.5 else None
    activities_list.append({
        "event_id": f"E{event_id:08d}",
        "student_id": stu,
        "class_id": cl,
        "activity_type": act,
        "timestamp": ts.isoformat(),
        "duration_sec": duration,
        "score": score
    })
    event_id += 1

df_activities = pd.DataFrame(activities_list)
df_activities.head()
```

```python
# ===========================================
# ✅ Cell 11 - Save all data with proper datetime format and UTF-8 encoding
# ===========================================
import pandas as pd
import json

def fix_datetime_columns(df):
    """
    Chuẩn hóa các cột thời gian:
    - Chuyển sang kiểu datetime
    - Xuất thành chuỗi ISO 8601: YYYY-MM-DDTHH:MM:SS
    """
    for col in df.columns:
        if any(k in col.lower() for k in ['date', 'time', 'dob', 'ngay', 'gio']):
            df[col] = pd.to_datetime(df[col], errors='coerce')
            df[col] = df[col].apply(lambda x: x.strftime('%Y-%m-%dT%H:%M:%S') if pd.notna(x) else '')
    return df

def clean_and_save(df, name):
    """
    Lưu mỗi bảng ở 2 định dạng:
    - CSV: cho phân tích bằng Spark, Power BI, Excel
    - JSON: cho import vào MongoDB
    """
    df = fix_datetime_columns(df)
    
    csv_file = f"{name}.csv"
    json_file = f"{name}.json"
    
    # Lưu CSV (UTF-8-SIG cho Excel hiển thị tiếng Việt đúng)
    df.to_csv(csv_file, index=False, encoding="utf-8-sig")
    
    # Lưu JSON (dạng record để import vào MongoDB)
    df.to_json(json_file, orient="records", force_ascii=False, date_format='iso', indent=2)
    
    print(f"💾 Saved: {csv_file} & {json_file}")

# ======= Save all tables =======
clean_and_save(df_students, "students")
clean_and_save(df_teachers, "teachers")
clean_and_save(df_courses, "courses")
clean_and_save(df_classes, "classes")
clean_and_save(df_enroll, "enrollments")
clean_and_save(df_sessions, "sessions")
clean_and_save(df_attendance, "attendance")
clean_and_save(df_grades, "grades")

# File hoạt động học tập (log lớn) — lưu parquet
df_activities = fix_datetime_columns(df_activities)
df_activities.to_parquet("activities.parquet", index=False)
df_activities.to_json("activities.json", orient="records", force_ascii=False, indent=2)
print("💾 Saved: activities.parquet & activities.json")

print("\n✅ All files successfully saved (CSV + JSON + Parquet) with ISO datetime format!")
```
